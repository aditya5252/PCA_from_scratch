```python
from src.import_file import *
from src import data_read
from src import high_dim_feature_with_labels
from src import PCA_file
from src import LDA_self
from sklearn.decomposition import PCA
'''Check results with PCA inbuilt'''
img_list,img_targets=data_read.get_data_(train=True)
N=len(img_targets)
X,Y=high_dim_feature_with_labels.norm_highx_y_(img_list,img_targets)

'K is the lower dimension'
for K in range(2,20):
    libpca = PCA(n_components=K)
    X_low_lib=libpca.fit_transform(X)
    mypca=PCA_file.PCA_scratch(n_components=K)
    X_low_my=mypca.fit_predict(X)
    print(f'{K}components mine')
    LDA_self.FiscLDA(X_low_my,Y)
    print(f'{K}components library')
    LDA_self.FiscLDA(X_low_lib,Y)
```
